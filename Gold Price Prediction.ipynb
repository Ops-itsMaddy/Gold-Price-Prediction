**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

**Data Collection and Processing**

```python
#Loading the csv data into the Pandas DataFrame
gold_data = pd.read_csv("/content/gld_price_data.csv")
```

```python
# Print first five rows of the dataframe
gold_data.head()
```

```python
# Print last five rows of the dataframe
gold_data.tail()
```

```python
# Number of rows and columns
gold_data.shape
```

```python
# Getting some basic information about the data
gold_data.info()
```

```python
# Checking the number of missing values
gold_data.isnull().sum()
```

```python
# Getting the statistical measures about the data
gold_data.describe()
```

**Correlation:**
1. Positive Correlation
2. Negative Correlation

```python
correlation = gold_data.drop('Date', axis=1).corr()
```

```python
# Constructing a heatmap to understand the correlation
plt.figure(figsize = (8,8))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size' : 8}, cmap = 'Reds')
```

```python
# Correlation values of GLD
print(correlation["GLD"])
```

```python
# Checking the Distribution of the GLD price
sns.displot(gold_data['GLD'], color = 'blue')
```

**Splitting the Features and the Target**

```python
X = gold_data.drop(columns = ['Date','GLD'], axis = 1)
Y = gold_data['GLD']
```

```python
print(X)
print(Y)
```

**Splitting into Training and Testing Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state= 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

Random Forest Regressor

```python
model = RandomForestRegressor(n_estimators = 100)
```

```python
model.fit(X_train, Y_train)
```

**Model Evaluation**

```python
# Prediction on Test Data
X_test_prediction = model.predict(X_test)
```

```python
# R Squared Error
error_score = metrics.r2_score(Y_test, X_test_prediction )
```

```python
print("R Squared Error :", error_score)
```

**Compare the Actual Values and the Predicted Values in a Plot**

```python
Y_test = list(Y_test)
```

```python
plt.plot(Y_test, color = 'Blue', label = "Actual Value")
plt.plot(X_test_prediction, color = 'Red', label = "Predicted Value")
plt.title("Actual Price vs Predicted Price")
plt.xlabel("Number of values")
plt.ylabel("GLD Price")
plt.legend()
plt.show()
```
